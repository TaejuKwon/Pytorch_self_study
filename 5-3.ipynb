```python
import torch
import torch.nn.functional as F
```

```python
torch.manual_seed(1)
```

```python
z = torch.FloatTensor([1, 2, 3])
```

```python
hypothesis = F.softmax(z, dim = 0)
print(hypothesis)
```

```python
hypothesis.sum()
```

```python
z = torch.rand(3, 5, requires_grad = True)
```

```python
hypothesis = F.softmax(z, dim = 1)    # 각 샘플에 대해서 소프트맥스 함수를 적용해야 하므로 두번째 차원(dim = 1)에 대해서 함수를 적용
print(hypothesis)
```

```python
y = torch.randint(5, (3, )).long()    # 임의의 레이블 만들기
print(y)
```

```python
y_one_hot = torch.zeros_like(hypothesis)
y_one_hot.scatter_(1, y.unsqueeze(1), 1)    # 0, 2, 1 / unsqueeze에서 (3,) -> (3 X 1)
```

```python
cost = (y_one_hot * -torch.log(hypothesis)).sum(dim = 1).mean()
print(cost)
```

```python
F.log_softmax(z, dim = 1)
```

```python
# cost

(y_one_hot * -torch.log(hypothesis)).sum(dim = 1).mean()
```

```python
# Also cost (No need to use one-hot encoding in y)

F.nll_loss(F.log_softmax(z, dim = 1), y)
```

```python
# Also cost

F.cross_entropy(z, y)
```
